## Exercice 2
Implémenter les versions itérative et récursive de la recherche du minimum dans une liste de taille minimale 1.



```python
def minimum_rec(l):
    """ list -> int or float or string
        renvoie l'élément minimum d'une liste
    """

    if len(l) == 1:
        return l[0]

    else:
        return min(l[0], minimum_rec(l[1:]))


    


def minimum_iter(l):
    """ list -> int or float or string
        renvoie l'élément minimum d'une liste
    """
    min = l[0]
    for x in l:
        if x < min:
            min = x

    return min
   
    
    
print(minimum_rec(list(range(3,12))))
print(minimum_iter(list(range(4,12))))
```

## Exercice 4
Ecrire une fonction récursive <code>somme_rec(l)</code> qui calcule la somme des éléments de la liste l. On supposera que la somme des éléments d'une liste vide est 0.

```python
def somme_rec(l):
    """
    list -> float
    renvoie la somme des éléments de la liste passée en paramètre
    """

    if l == []:
        return 0

    if len(l) == 1:
        return l[0]

    return l[0] + somme_rec(l[1:])
    
    
l = list(range(4))
somme_rec(l)
```

## Exercice 5 : suite de Syracuse
Soit $u_n$ la suite d'entiers définie par :
$$ 
u_{n+1} = \left\{ 
\begin{array}{cc}
u_n/2 & si \;u_n \;est\; pair\\
3 \times u_n + 1 & sinon
\end{array} \right. $$
avec $u_0$ un entier quelconque supérieur à 1.

Ecrire une fonction récursive <code>syracuse(u_0)</code> qui renvoie les valeurs successives de la suite de Syracuse de premier terme <code>u_0</code> tant que $u_n$ est plus grand que 1. 
Tester ensuite la fonction avec différents premiers termes $u_0$.

```python
[2] + [3]
```

```python
def syracuse(u):
    """
    int -> list
    renvoie les termes de la suite de Syracuse de premier terme 
    donné en paramètre
    """
    if u == 1:
        return[1]
    
    else:
        if u % 2 == 0:
            u = u//2
        else:
            u = (3*u) + 1

        return [u] + syracuse(u)


import numpy as np
import matplotlib.pyplot as plt   
def syracuse_len(u):
    """
    int -> int
    renvoie la longueur de la suite de Syracuse
    de premier terme donné en paramètre
    """
    return len(syracuse(u))

l = []
t = np.arange(1,10000)
for i in t:
    l = l + [syracuse_len(i)]   

plt.plot(np.log(t), np.log(l))
#plt.plot(t, l)
print(syracuse(13))
```

La conjecture de Syracuse affirme que, quelle que soit la valeur de $u_0$, il existe toujours un indice $n$ dans la suite tel que que $u_n = 1$. Cette conjecture défie toujours les mathématiciens.

## Exercice 6 
Un palindrome est un mot dont les lettres, lues de droites à gauche sont les mêmes que celles lues de gauche à droite. Les mots "radar", "elle", "été", "ici" sont des palindromes. "raper" n’est pas un palindrome.
Ecrire une fonction récursive qui teste si un mot est un palindrome.

```python
def check_palindrome(s):
    """
    str -> bool
    renvoie True si le mot donné en paramètre est un palindrome,
    renvoie False sinon
    """

    if s[0] != s[-1]:
        return False
    if len(s) == 1:
        return True
    return check_palindrome(s[1:-1])
    
    

check_palindrome("rarr")
```

## Exercice 7
Ecrire une fonction récursive <code>retourner(mot)</code> prenant en argument une chaine de caractères et renvoyant la même chaine écrite de droite à gauche. Par exemple, <code>retourner('robert')</code> renverra 'trebor'.

```python
def retourner(s):
    """
    str -> str
    renvoie la chaine entrée en paramètre écrite de droite à gauche
    """

    if len(s) == 1:
        return s

    else:
        return retourner(s[1:]) + s[0]
    
    
retourner("oiseau")
```

```python
def nombre_de_bits_1(n):
    
    
    if type(n) == int:
        n = "{0:b}".format(n)
    if len(n) == 1:
        return int(n[0])
    return int(n[0]) + nombre_de_bits_1(n[1:])

print(nombre_de_bits_1(234))


```

## Exercice 10
Ecrire une fonction récursive <code>appartient(v, t, i)</code> prenant en argument une valeur <code>v</code>, un tableau <code>t</code> et un entier <code>i</code> et renvoyant <code>True</code> si <code>v</code> apparait dans <code>t</code> entre l'indice <code>i</code> (inclus) et <code>len(t)</code> (exclu), et <code>False</code> sinon.
On supposera que <code>i</code> est toujours compris entre 0 et <code>len(t)</code>.

```python
def appartient(v, t, i):
    """
    int(ou float ou str), list, int -> bool
    renvoie True si la valeur v apparait dans le tableau t à l'indice i ou après 
    l'indice i
    """
    assert 0 <= i <= len(t), "i est trop grand"

    if t[i] == v:
        return True
    if i == len(t)-1:
        return False

    return appartient(v, t, i+1)
   

v = 7
t = list(range(1,10))
i = 4

print(v)
print(t)
print(i)

appartient(v,t,i)
```
