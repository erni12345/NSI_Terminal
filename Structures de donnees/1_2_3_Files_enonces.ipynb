```python
class Node:
    def __init__(self,tete,queue):
        self.car=tete
        self.cdr=queue
        
class File:
  
  def __init__(self):
    self.head = None
    self.last = None

  def is_empty(self):
    return self.head == None

  def push(self,el):
    c = Node(el, None)

    if self.head is None:
      self.head = c

    else:
      self.last.cdr = c
    
    self.last = c
    

  def pop(self):

    head = self.head
    if head is not None:
      self.head = head.cdr
    return head

      


    
```

Tests

```python
F=File()
F.push(1)
F.push(2)
print(F.head.car)
print(F.last.car)
F.pop()
print(F.head.car)
```
