<li> Ecrire la fonction de parcours en largeur d'un arbre binaire, en suivant l'algorithme :

    Fonction parcours_largeur(arbre) 
    Données : arbre binaire
    f une file vide
    li une liste vide
    On met l’arbre dans la file
    Tant que la file n’est pas vide faire
        On défile la file dans une variable tmp 
        On ajoute tmp.valeur à la liste li
        Si tmp.gauche n’est pas vide alors
            On enfile tmp.gauche
        Si tmp.droit n’est pas vide alors 
            On enfile tmp.droit
    renvoyer liste li

 </li>
 <li>Ecrire le code qui crée l'arbre représenté dans l'exercice 4 de la feuille. </li>
 <li>Tester l'algorithme sur cet arbre</li>

</ol>


```python
class Node:
    def __init__(self, value, left, right):
        self.value = value
        self.left = left
        self.right = right


class Tree:

    def __init__(self):
        self.root = Node(None, None, None)

    
    def taille(self, current):
        if current is None:
            return 0
        else:
            return 1 + self.taille(current.right) + self.taille(current.left)    

    def hauteur(self, current, n=0):
        if current is None:
            return n
        else :
            return max(self.hauteur(current.right, n+1), self.hauteur(current.left, n+1))


class NodeF:
    def __init__(self,v,s):
        self.valeur=v
        self.suivant=s

class File:
  """structure de file"""
  def __init__(self):
    self.first=None
    self.last=None

  def is_empty(self):
    return self.first is None

  def push(self,el):
    c=NodeF(el,None)
    if self.is_empty():
      self.first=c
    else:
      self.last.suivant=c
    self.last=c

  def pop(self):
    assert not self.is_empty(), 'File vide'
    v=self.first.valeur
    self.first=self.first.suivant
    if self.first is None:
      self.last is None
    return v
    
def parcours_largeur(arbre):
    f = File()
    li = []

    f.push(arbre)
    while not f.is_empty():
      current = f.pop()
      li.append(current.value)
      if current.left is not None:
        f.push(current.left)
      if current.right is not None:
        f.push(current.right)

    return li


tree = Tree()

tree.root = Node(20, None, None)
tree.root.left = Node(5, Node(3, None, None), Node(12, Node(8, Node(6, None, None), None), Node(13, None, None)))
tree.root.right = Node(25, Node(21, None, None), Node(28, None, None))

root = tree.root

print(parcours_largeur(root))
```

```python
li = []

def parcours_prefixe(arbre):
    
    if arbre:
        li.append(arbre.value)
        parcours_prefixe(arbre.left)
        parcours_prefixe(arbre.right)

parcours_prefixe(root)
print(li)

li = []
def parcours_infixe(arbre):

    if arbre:
        parcours_infixe(arbre.left)
        parcours_infixe(arbre.right)
        li.append(arbre.value)

parcours_infixe(root)
print(li)


li = []

def parcours_suffixe(arbre):
    
    if arbre:
        parcours_suffixe(arbre.left)
        li.append(arbre.value)
        parcours_suffixe(arbre.right)

parcours_suffixe(root)
print(li)
```
