<h2><b><font color="orange"> DICTIONNAIRES</font></b></h2>
<hr color="orange">

<p>Nous allons étudier un autre type abstrait de données : les dictionnaires . 
On retrouve une structure qui ressemble, à première vue, beaucoup à un tableau (à chaque élément  on associe un indice de position).  
    
Mais au lieu d'associer chaque élément à un indice de position, dans un dictionnaire, on associe  chaque élément (on parle de valeur dans un dictionnaire) à une clef, on dit qu'un dictionnaire  contient des couples <code>clef:valeur</code> (chaque clé est associée à une valeur). 
</p>
<p>Voici les opérations que l'on peut effectuer sur le type abstrait dictionnaire : <ul> 
    <li><code>ajout</code> : on associe une nouvelle valeur à une nouvelle clé  </li>
    <li><code>modif </code>: on modifie un couple clé:valeur en remplaçant la valeur courante par une autre  valeur (la clé restant identique)  </li>
    <li><code>suppr</code> : on supprime une clé (et donc la valeur qui lui est associée)  </li>
    <li><code>rech</code> : on recherche une valeur à l'aide de la clé associée à cette valeur. </li> </ul>
</p>



<h2><font color="orange"> Première implémentation</font></h2>

1/ Construire une classe <code>dict1</code> qui n’aura qu’un seul attribut <code>Liste</code> : une liste vide.  Construire une méthode <code>ajout</code> qui prend comme argument une liste <code>[cle,valeur]</code> et qui ajoute  cette liste à l’attribut Liste. 

2/ Ecrire une méthode <code>affiche</code> permettant l’affichage comme ci-dessous. 
        
        L=dict1() 
        L.ajout(["cle1","val1"]) 
        L.ajout(["cle2","val2"])  
        L.affiche() 
        
        {cle1 : val1 ; cle2 : val2 ; } 
        
3/ Construire les méthodes <code>suppr, modif, rech </code>. Exemple d'utilisation :
        
        L.modif("cle1","Val1modif") 
        L.suppr("cle2") 
        L.affiche() ->  {cle1 : Val1modif ; }
        print(L.rech("cle1")) -> Val1modif 
        print(L.rech("cle2")) -> Clé non trouvée 



```python
class dict1:
  
  def __init__(self):
    self.keys = []
    self.items = []

  def ajout(self, lst):
    if lst[0] not in self.keys:
      self.keys.append(lst[0])
      self.items.append(lst[1])

  def affiche(self):
     print("{", end = "")
     for x in range(len(self.keys)):
       print(f"{self.keys[x]} : {self.items[x]}, ", end = "")
     print("}")

  
  def modif(self, key, item):
    try:
      index = self.keys.index(key)
      if index >= 0:
        self.items[index] = item
    except:
      self.ajout([key, item])
  
  def suppr(self, index):
    try:
      index = self.keys(index)
      self.keys.pop(index)
      self.items.pop(index)
    except:
      return "not in list"

  def rech(self, index):
    try:
      index = self.keys(index)
      return self.items[index]

    except:
      return "Pas dans la liste"




L=dict1() 
L.ajout(["cle1","val1"]) 
L.ajout(["cle2","val2"])  
L.affiche()
L.modif("cle1","Val1modif") 
L.affiche()
L.suppr("cle2")
print(L.rech("cle2"))
L.affiche()
```
